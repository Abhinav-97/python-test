# we have taken stock prices for 3 years

```python
data = get_history(symbol='SBIN',
                    start=date(2015,1,1), 
                    end=date(2017,12,30))
```

```python
data.head(15)
```

# get all the dates and then determine the week number for that year

```python
dates = data.index.values.tolist()
```

# Assign  a continous week number to get moving averages

```python
week = []
for i in range(len(dates)):
    week.append(int(dates[i].strftime("%V")))
```

```python
data["week"] = week
```

```python
week_bias = {'2015':0,
            '2016':52,
            '2017':104}
week_number  = []
years = []
for i in range(len(dates)):
    year = dates[i].strftime("%Y")
    years.append(year)
    week_number.append(week[i] + week_bias[year])
```

```python
data["week_number"] = week_number
```

```python
data["year"] = years
```

# Function to determine moving average

```python
def movingaverage(data, window_size):
    week_data = data.groupby("week_number")["Close"].mean().values
    window = np.ones(int(window_size))/float(window_size)
    return np.convolve(week_data, window, 'valid')
```

```python
(movingaverage(data, 4))
```

# Get Price Shock

```python
data["price_shock"] = (((abs(data["Close"] - data["Prev Close"])/data["Prev Close"])*100) > 2).astype(int)
```

```python
# 0 for negative 1 for positive
data["price_shock_direction"] = (data["Close"] - data["Prev Close"] > 0).astype(int)
```

# Get Volume Shock

```python
data["volume_shock"] = ((abs((data["Volume"].shift(1) - data["Volume"])/data["Volume"].shift(1))*100)  > 10).astype(int)
```

# Get Price Shock without Volume shock

```python
# data["price_and_volume_shock"] = data["volume_shock"] & data["price_shock"]
data["not_volume_shock"]  = (~(data["volume_shock"].astype(bool))).astype(int)
data["price_shock_without_volume_shock"] = data["not_volume_shock"] & data["price_shock"]
```
